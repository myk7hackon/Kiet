```python
def minkowskiDistance(instance1, instance2, length):
    distance = 0
    for x in range(length):
        distance += pow((instance1[x] - instance2[x]), length+1)
    return distance**(1/length)
```

```python
def getNeighbors(trainingSet, testInstance, k):
    distances = []
    length = len(testInstance)-1
    #print(length)
    for index,row in trainingSet.iterrows():
        
        train_input=list(row)
        #print(train_input)
        dist = minkowskiDistance(train_input, testInstance, length)
        distances.append((row, dist))
    #print(distances)
    distances.sort(key=operator.itemgetter(1))
    neighbors = []
    for x in range(k):
        neighbors.append(distances[x][0])
    return neighbors
```

```python
def getResponse(neighbors):
    classVotes = {0.0:0,1.0:0}
    for x in neighbors:
        response = x.Survived
        if response in classVotes:
            classVotes[response] += 1
        else:
            classVotes[response] = 1
    sortedVotes = sorted(classVotes.items(), key=operator.itemgetter(1), reverse=True)
    return sortedVotes[0][0]
```

```python
def getAccuracy(test_data, predictions):
    correct = 0
    x=0
    for index,row in test_data.iterrows():
        if row.Survived == predictions[x]:
            correct += 1
        x=x+1
    return (correct/float(len(predictions))) * 100.0
```
