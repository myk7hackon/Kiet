# Handling missing age...

## Strategies?

<li> Drop data?
<li> Set all the NaN ages to 0(for continuous) and for categorical data we can create some categories.
<li> Use some average values?

## So who's the winner?

```python
avgAge=df.Age.mean()
```

```python
print(avgAge)
```

```python
df.Age=df.Age.fillna(value=avgAge)
```
